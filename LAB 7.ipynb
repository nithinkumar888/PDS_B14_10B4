```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("/content/archive (25).zip")

# Extract years and population data
years = [int(col.split(" ")[0]) for col in data.columns if "Population" in col and col.split(" ")[0].isdigit()]
# Calculate the sum of population for each year across all countries
population_by_year = data[data.columns[data.columns.str.contains('Population') & data.columns.str.split(' ').str[0].str.isdigit()]].sum()

# Plot line chart
plt.figure(figsize=(10,6))
plt.plot(years, population_by_year, marker='o', linestyle='-', color='b', linewidth=2)

# Add labels and title
plt.xlabel("Year", fontsize=12)
plt.ylabel("World Population", fontsize=12)
plt.title("World Population Growth Over Decades", fontsize=14, fontweight="bold")

# Format y-axis (convert to billions for readability)
# Create custom labels for the y-axis in billions
y_labels = [f"{int(pop/1e9)}B" for pop in population_by_year]
plt.xticks(years) # Set x-axis ticks to the years
plt.yticks(population_by_year, y_labels) # Set y-axis ticks and labels

# Show grid
plt.grid(True, linestyle="--", alpha=0.6)

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv("/content/archive (26).zip")

# Convert Date column to datetime format
data["Date"] = pd.to_datetime(data["Date"])

# Plot line chart
plt.figure(figsize=(12,6))
plt.plot(data["Date"], data["Close"], color="blue", linewidth=2, marker="o", markersize=4)

# Formatting x-axis for dates
plt.gcf().autofmt_xdate()  # Rotate date labels automatically

# Add labels and title
plt.xlabel("Date", fontsize=12)
plt.ylabel("Stock Price (USD)", fontsize=12)
plt.title("Apple (AAPL) Stock Price Trend", fontsize=14, fontweight="bold")

# Add grid
plt.grid(True, linestyle="--", alpha=0.6)

# Show plot
plt.show()
```
